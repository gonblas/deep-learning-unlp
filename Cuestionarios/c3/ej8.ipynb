```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from sklearn.preprocessing import StandardScaler
from sklearn.utils import shuffle
from tensorflow.keras.utils import to_categorical

# Fijar semilla para reproducibilidad
tf.random.set_seed(1)

# Cargar datos (asumiendo que el archivo está en el mismo directorio)DATOS_DIR = '../../Datos/'
DATOS_DIR = '../../Datos/'
data = pd.read_csv(DATOS_DIR + 'vinos.csv', sep=';')


# Separar características y etiquetas
X = data.iloc[:, 1:].values   # 13 atributos
y = data.iloc[:, 0].values    # tipo de vino (1, 2, 3)

# Normalización con media y desvío estándar
scaler = StandardScaler()
X_norm = scaler.fit_transform(X)

# Codificar etiquetas en one-hot (1→0, 2→1, 3→2)
y_cat = to_categorical(y - 1)

# Parámetros del entrenamiento
lr = 0.05
epochs = 50
batch_size = 20

def crear_y_entrenar_modelo():
    # Definir modelo secuencial
    model = models.Sequential([
        layers.Input(shape=(13,)),
        layers.Dense(10, activation='tanh'),  
        layers.Dense(3, activation='sigmoid')
    ])
    
    # Compilar
    opt = optimizers.SGD(learning_rate=lr)  
    model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
    
    # Entrenar
    history = model.fit(X_norm, y_cat, epochs=epochs, batch_size=batch_size, verbose=0)
    
    # Evaluar
    loss, acc = model.evaluate(X_norm, y_cat, verbose=0)
    print(f"Accuracy final: {acc:.4f}")
    return acc

# Entrenar con ambas funciones de activación
acc_sigmoid = crear_y_entrenar_modelo()
acc_relu = crear_y_entrenar_modelo()
```
