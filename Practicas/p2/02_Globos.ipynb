## Ejercicio 2

El archivo **globos.csv** contiene el registro de 16 intentos para inflar globos. Los atributos registrados fueron color del globo, el tamaño, si se estira o no y si la acción de inflarlo fue realizada por un adulto o por un niño. En cada registro se anotó si el globo pudo ser inflado o no.

Numerice los atributos de la siguiente forma:
- Color = 1 si es rojo y 0 si es amarillo
- SeEstira? = 1 para Si y 0 para No
- Edad = 1 si es adulto y 0 si es Nene
- Tamanio = 0 si es Chico, 0.5 si es mediano y 1 si es grande

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

DATOS_DIR   = '../../Datos/p2/'
data = pd.read_csv(DATOS_DIR + 'globos.csv')


# Color
NewColorColumn = pd.get_dummies(data['Color'], prefix= 'Color', drop_first=True).astype(int)

data = pd.concat([NewColorColumn, data], axis=1)
data.drop(['Color'],axis=1, inplace=True)


# Tamanio
tamanioMap ={
	"Tamanio": {
		"Chico": 0,
		"Mediano": 0.5,
		"Grande": 1
	}
}
data.replace(tamanioMap, inplace=True)

# SeEstira?
NewSeEstiraColumn = pd.get_dummies(data['Se_estira?'], prefix= 'Se_estira?', drop_first=True).astype(int)
data = pd.concat([NewSeEstiraColumn, data], axis=1)
data.drop(['Se_estira?'],axis=1, inplace=True)

# Edad
NewEdadColumn = pd.get_dummies(data['Edad'], prefix= 'Edad', drop_first=False).astype(int)
NewEdadColumn.drop('Edad_Nene', axis=1, inplace=True)
data = pd.concat([NewEdadColumn, data], axis=1)
data.drop(['Edad'],axis=1, inplace=True)

# Inflado?: Si bien no lo pide lo voy a necesitar para entrenar el perceptron
NewInfladoColumn = pd.get_dummies(data['Inflado?'], prefix= 'Inflado?', drop_first=True).astype(int)
data = pd.concat([NewInfladoColumn, data], axis=1)
data.drop(['Inflado?'],axis=1, inplace=True)

data
```

```python
perceptrons = [
    {
        "w_color": 0.050,
        "w_tamanio": 0.006,
        "w_se_estira": -0.068,
        "w_edad": -0.047,
        "b": 0.005
    },
    {
        "w_color": -0.494,
        "w_tamanio": -0.003,
        "w_se_estira": 0.722,
        "w_edad": 0.489,
        "b": -0.4943
    },
    {
        "w_color": 0.06,
        "w_tamanio": 0.004,
        "w_se_estira": 0.1278,
        "w_edad": 0.021,
        "b": -0.1132
    },
    {
        "w_color": 0.296,
        "w_tamanio": 0.026,
        "w_se_estira": -0.576,
        "w_edad": -0.287,
        "b": 0.371
    },
    {
        "w_color": 0.994,
        "w_tamanio": 0.006,
        "w_se_estira": -1.466,
        "w_edad": -0.989,
        "b": 0.988
    }
]

for i, p in enumerate(perceptrons):
    incorrect = 0
    for idx, row in data.iterrows():
        z = (p["w_color"] * row["Color_Rojo"] +
			p["w_tamanio"] * row["Tamanio"] +
			p["w_se_estira"] * row["Se_estira?_Si"] +
			p["w_edad"] * row["Edad_Adulto"] -
			p["b"])
        a = 1 if (z >= 0) else 0
        data.at[idx, f'Perceptron_{i+1}'] = a
        if a != row["Inflado?_Si"]:
            incorrect += 1
    print(f"Perceptron {i+1}: {incorrect} errores de {len(data)} ejemplos.")

data # Ninguno clasifica todo correctamente, pero el que menos errores tiene es el perceptron 4 con 6 errores.
```
